```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np

traj_noAB = md.load('ATP_IDE.dcd', top='system_ATP_IDE.pdb')
traj_withAB = md.load('ATP_IDE_AB.dcd', top='system_ATP_IDE_AB.pdb')

noAB = traj_noAB.topology.select('resid 375 to 862')
withAB = traj_withAB.topology.select('resid 375 to 862')

_noAB = traj_noAB.atom_slice(noAB)
_withAB = traj_withAB.atom_slice(withAB)

rmsf_noAB= md.rmsf(_noAB, _noAB, 0)
rmsf_withAB= md.rmsf(_withAB, _withAB, 0)

X_noAB= list(noAB) 
Y_noAB= list(rmsf_noAB)
X_withAB= list(withAB) 
Y_withAB= list(rmsf_withAB)

plt.plot(X_noAB,Y_noAB, color='green', label = "ATP-IDE (300 K)") # "ATP-IDE (315.15 K)"; "ATP-IDE (321.15 K)"; 
plt.plot(X_withAB,Y_withAB, color='black', label = "ATP-IDE-A (300 K)") # "ATP-IDE (315.15 K)"; "ATP-IDE (321.15 K)"; 

plt.axvspan(6143,6164,color="blue", alpha=0.8, label ="LYS384") 
plt.axvspan(6165,6176,color="red", alpha=0.8, label ="ASP385")
plt.axvspan(6199,6213,color="violet", alpha=0.8, label = "GLU387")

plt.axvspan(8583,8604,color="brown", alpha=0.8, label ="LYS530")
plt.axvspan(8615,8628,color="orange", alpha=0.8, label ="LYS532") 

plt.axvspan(9306,9319,color="cyan", alpha=0.8, label ="SER576")
plt.axvspan(9579,9599,color="yellow", alpha=0.8, label ="SER593")

plt.axvspan(13996,14018,color="pink", alpha=0.8, label ="LYS858")
plt.axvspan(14037,14048,color="olive", alpha=0.8, label ="LYS860")


plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left', fontsize=18)

plt.rcParams["figure.figsize"] = (25,5)

plt.ylim(0,10)

plt.xlabel("Atom index", fontweight='normal', fontsize=18)
plt.ylabel("RMSF ($\AA$)", fontweight='normal', fontsize=18)

plt.savefig('All', dpi=200, bbox_inches='tight')
```
